**Distribution Plots**

```
import seaborn as sns
```

```
%matplotlib inline
```

```
tips = sns.load_dataset('tips')
```

```
tips.head()
```

```
#Distribution plot
sns.displot(tips['total_bill'],kde=False,bins=30)
```

```
#Joint Plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```
sns.rugplot(tips['total_bill'])
```

```
sns.kdeplot(tips["total_bill"])
```

```
import numpy as np
```

**Matrix Plots**

```
flights = sns.load_dataset('flights')
```

```
flights.head()
```

```
# for creating heat map the first step is to convert your data into matrix form which means the index name and column name match us so that the cell value
#indicate something that is relevant to both of those names.
tc = tips.corr()
tc
```

```
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```
#Heat map for flights
fc = flights.pivot_table(index='month',columns='year',values='passengers')
fc
```

```
#simple
sns.heatmap(fc)
#more attributes
#sns.heatmap(fc,cmap='magma',linecolor='white',linewidths=1)
```

```
#Cluster Map
#sns.clustermap(fc)
#Cluster Map with more attributes
sns.clustermap(fc,cmap='coolwarm',standard_scale=1)
```

**Regression Plot**

```
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100},)
```

```
#Now if we want to see data more clearly so that data for male and female get separated in culumns
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```
#aspect and size Ratio's
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,size=8)
```

**Seaborn Grid Plots**

```
iris=sns.load_dataset('iris')
```

```
iris.head()
```

```
iris['species'].unique()
```

```
#First we clreate a PairPlot
sns.pairplot(iris,hue='species')
```

```
import matplotlib.pyplot as plt
```

```
#In pairGrid mechanism of seaborn we have a lot more control over actual plot.we can do more customization.
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```
#If we want different types of plots in a grid
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```
#FacetGrid
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```
#scatter Plot
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

**Color And Style**

```
#set backgrount of grid as white/darkgrid/ticks/whitegrid
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
#is we want to remove spine
#sns.despine()
sns.despine(left=True,bottom=True)
```

```
#Change Size and Aspect Ratio
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```
#Scale and Context
#If we are making figure for Posters
sns.set_context('poster')
sns.countplot(x='sex',data=tips)

```

```
#we can also set font size rather than default
sns.set_context('poster',font_scale=3)#it means 3 times default
sns.countplot(x='sex',data=tips)
```

```
#palette and colors
#see decumentation for matplotlib color palette(seaborn use matplotlib color palette)
sns.set_context('notebook')#if you not convert it into notebook it will display in poster contex
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```
