```
import numpy as np
import pandas as pd
```

```
labels = ['a','b','c']
my_data = [10,20,30]
my_array = np.array(my_data)
d = {'a':10,'b':20,'c':30}
```

```
#it is similar to numpy arrays but with indexes
pd.Series(data = my_data)
```

```
#we can also have indexes of our choice.
#we can cal this data by using these labels
pd.Series(data = my_data,index=labels)
```

```
#both are similar
pd.Series(my_data,labels)
```

```
#we can pass both a list and numpy arrays 
pd.Series(my_array,labels)
```

```
pd.Series(d)
```

```
#series can hold any type of data objects
pd.Series(labels)
```

```
#it can also hold instances of functions.
pd.Series([sum,print,len])
```
